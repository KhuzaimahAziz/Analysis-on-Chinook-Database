```python
q='''select alb.title Album_Title, ar.name Artist_Name,
        count(il.track_id) No_purchases
        from artist ar
        left join album alb
        on
        alb.artist_id=ar.artist_id
        inner join track t
        on
        t.album_id=alb.album_id
        inner join invoice_line il
        on
        t.track_id=il.track_id
        group by 1
        order by 3 desc'''
read_sql(q)
```

```python
q='''select c.first_name||' '||c.last_name Customer_Name,c.Country,
     sum(inv.total) Purchases from customer c
     inner join invoice inv
     on
     c.customer_id=inv.customer_id
     group by 1'''
read_sql(q)
```

```python
q='''with Total_Purchase as (select c.first_name||' '||c.last_name Customer_Name,c.Country,
     sum(inv.total) Purchases from customer c
     inner join invoice inv
     on
     c.customer_id=inv.customer_id
     group by 1)
     select Customer_Name,Country, max(Purchases) Top_Buyers from Total_Purchase
     group by 2
     order by 3 desc'''
read_sql(q)
```

```python
q='''select c.first_name||' '||c.last_name Customer_Name,c.Country,
     count(inv.invoice_id)number_of_purchases,
     sum(inv.total) Purchases,
     case
     when sum(inv.total) < 40 then 'Small Spender'
     when sum(inv.total) >100 then 'Big Spender'
     else 'Regular'
     end as Spender
     from customer c
     inner join invoice inv
     on
     c.customer_id=inv.customer_id
     group by 1'''
read_sql(q)
```

```python
import matplotlib.pyplot as plt
```

```python
q='''with US_Genre_Sales as(select g.name Genre_Name, count(t.track_id) Total_Sales from genre g
     inner join  track t
     on
     g.genre_id=t.genre_id
     inner join invoice_line il
     on
     t.track_id=il.track_id
     inner join  invoice i
     on
     il.invoice_id=i.invoice_id
     inner join customer c
     on 
     c.customer_id=i.customer_id
     where c.country='USA'
     group by 1
     order by 2 desc
     )
     select Genre_Name,Total_Sales,
     cast(Total_Sales as float)/(select sum(Total_Sales)from US_Genre_Sales)*100 Percentage
     from US_Genre_Sales
     '''
df=read_sql(q)
```

```python
genre=df.set_index('Genre_Name')
```

```python
for i,gen in enumerate(genre.index)
    label=str(round(genre.loc[i, 'Percentage'],2))+'%'
    x=genre.loc[i, 'Total_Sales']
    plt.annotate(label, x, i)
```

```python
genre['Total_Sales'].plot.barh(xlim=(0,700))
for i,gen in enumerate(genre.index):
    label=str(round(genre.loc[gen, 'Percentage'],1))+'%'
    x=genre.loc[gen, 'Total_Sales']
    plt.annotate(label, (x+20, i-0.05))
    plt.ylabel(' ')
```
